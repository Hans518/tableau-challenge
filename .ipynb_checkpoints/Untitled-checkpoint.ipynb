```python
import os
import glob
import pandas as pd
os. getcwd()
#os.chdir('JC-data')
```

```python
extension = 'csv'
all_filenames = [i for i in glob.glob('*.{}'.format(extension))]
```

```python
print(all_filenames)
```

```python
#combine all files in the list
combined_csv = pd.concat([pd.read_csv(f) for f in all_filenames ])
#export to csv
combined_csv.to_csv( "combined_csv.csv", index=False, encoding='utf-8-sig')
```

```python
data_to_load = "combined_csv.csv"
bike_data_df = pd.read_csv(data_to_load)
```

```python
bike_data_df = bike_data_df.rename(columns={'gender': 'gender_id'})
```

```python
bike_data_df
```

```python
gender_df = pd.DataFrame({'gender_id': [0,1,2],
                         'gender': ['unknown', 'male', 'female']})
gender_df
```

```python
citibike_data = pd.merge(bike_data_df, gender_df, on=["gender_id", "gender_id"])
citibike_data = citibike_data.drop(columns = ["gender_id"])
citibike_data
```

```python
citibike_data['starttime'] = pd.to_datetime(citibike_data['starttime'])
citibike_data['stoptime'] = pd.to_datetime(citibike_data['stoptime'])
```

```python
citibike_data
```

```python
stations = citibike_data['start station name'].unique()
```

```python
months_start = citibike_data.groupby(['start station name', pd.Grouper(key='starttime', freq='M')]).count()
months_stop = citibike_data.groupby(['end station name', pd.Grouper(key='stoptime', freq='M')]).count()
```
